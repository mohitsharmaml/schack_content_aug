```python
from PIL import Image, ImageDraw, ImageFont, ImageTk, features, ImageFile, ImageFilter, ImageChops
# Check if libraqm loaded (for proper indic language rendering)
print (features.check('raqm'))
```

```python
# Scraping shayari content from www.sanjayjangam.com
import requests
from bs4 import BeautifulSoup

def get_shayari(URL = "https://www.sanjayjangam.com/hindi/broken-heart-shayari-in-hindi"):
    
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    shy = soup.find_all(class_='wp-block-quote')
    shayari = []
    for x in list(shy):
        shayari.append(str(x).replace('<br/>',' ').replace('<blockquote class="wp-block-quote">','')
                       .replace('</blockquote>','').replace('<p>','').replace('</p>','').strip())
    return shayari
```

```python
shayari = get_shayari()
```

```python
# Captioning AI generated images with random quotes
import glob
import textwrap
import numpy as np

def to_image(shayari, in_path, out_path, font_path):
    cnt = 0
    for imgp in glob.glob(f"{in_path}/*.png"):
        cnt += 1
        img = Image.open(imgp)
        width, height = img.size

        text = shayari[int(np.random.rand(1)*len(shayari))]
        font_size = 25
    
        para = textwrap.wrap(text, width=24)

        MAX_W, MAX_H = img.size
        im = Image.new('RGBA', (MAX_W, MAX_H), (0, 0, 0, 0))
        draw = ImageDraw.Draw(im)
        font = ImageFont.truetype(f"{font_path}", font_size)

        current_h, pad = 150, 10
        for line in para:
            w, h = draw.textsize(line, font=font)
            draw.text(((MAX_W - w) / 2, current_h), line, fill='black', font=font)
            current_h += h + pad

        im_b = im.filter(ImageFilter.GaussianBlur(radius=2))
        draw = ImageDraw.Draw(im_b)
        current_h, pad = 150, 10
        for line in para:
            w, h = draw.textsize(line, font=font)
            draw.text(((MAX_W - w) / 2, current_h), line, fill='white', font=font)
            current_h += h + pad
    
        im_b = Image.composite(img, im_b, mask=ImageChops.invert(im_b))
        im_b.save(f'{out_path}/output-{cnt}.png')
```

```python
to_image(shayari, './imgs/heartbreak', './output', "./fonts/NotoSerifDevanagari-Medium.ttf")
```

```python
# Converting AI generated videos to short video status with music
from moviepy.editor import *
from moviepy.video.fx.all import crop
import math

def to_video(shayari, in_path, out_path, font_path, song_path):
    img = Image.open(f'{in_path}')
    width, height = img.size

    text = shayari[int(np.random.rand(1)*len(shayari))]
    font_size = 25
    
    para = textwrap.wrap(text, width=24)

    MAX_W, MAX_H = img.size
    im = Image.new('RGBA', (MAX_W, MAX_H), (0, 0, 0, 0))
    draw = ImageDraw.Draw(im)
    font = ImageFont.truetype(f"{font_path}", font_size)

    current_h, pad = 150, 10
    for line in para:
        w, h = draw.textsize(line, font=font)
        draw.text(((MAX_W - w) / 2, current_h), line, fill='black', font=font)
        current_h += h + pad

    im_b = im.filter(ImageFilter.GaussianBlur(radius=2))
    draw = ImageDraw.Draw(im_b)
    current_h, pad = 150, 10
    for line in para:
        w, h = draw.textsize(line, font=font)
        draw.text(((MAX_W - w) / 2, current_h), line, fill='white', font=font)
        current_h += h + pad
    
    im_b = Image.composite(Image.new('RGBA', (MAX_W, MAX_H), (0, 0, 0, 0)),im_b, mask=ImageChops.invert(im_b))
    im_b.save('temp.png')
    
    def zoom_in_effect(clip, zoom_ratio=0.02):
        def effect(get_frame, t):
            img = Image.fromarray(get_frame(t))
            base_size = img.size

            new_size = [
                math.ceil(img.size[0] * (1 + (zoom_ratio * t))),
                math.ceil(img.size[1] * (1 + (zoom_ratio * t)))
            ]

            new_size[0] = new_size[0] + (new_size[0] % 2)
            new_size[1] = new_size[1] + (new_size[1] % 2)

            img = img.resize(new_size, Image.LANCZOS)

            x = math.ceil((new_size[0] - base_size[0]) / 2)
            y = math.ceil((new_size[1] - base_size[1]) / 2)

            img = img.crop([
                x, y, new_size[0] - x, new_size[1] - y
            ]).resize(base_size, Image.LANCZOS)

            result = np.array(img)
            img.close()

            return result

        return clip.fl(effect)

    images = [f'{in_path}']

    slides = []
    for n, url in enumerate(images):
        slides.append(ImageClip(url).set_fps(24).set_duration(15))
        slides[n] = zoom_in_effect(slides[n], 0.02)

    video = concatenate_videoclips(slides)
    clip = video.subclip(0, 15)
  
    audioclip = AudioFileClip(f"{song_path}").subclip(0, 15)
  
    videoclip = clip.set_audio(audioclip)

    (w, h) = videoclip.size
    cropped_clip = crop(videoclip, width=w*9/16, height=h, x_center=w/2, y_center=h/2)
    
    cat = (ImageClip("temp.png").set_duration(clip.duration).set_position(lambda t: ('center', - h/2*(t/100))))

    cropped_clip = CompositeVideoClip([cropped_clip, cat])

    cropped_clip.write_videofile(f"{out_path}", codec='mpeg4', audio_codec='aac', 
                         temp_audiofile='temp-audio.m4a', remove_temp=True)
```
